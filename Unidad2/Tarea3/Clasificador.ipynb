## Se importan los datos del dataset

```python
import numpy as np
import pandas as pd

# cargar el dataset
data = pd.read_csv('./Dataset/spam.csv', encoding='ISO-8859-1')

# eliminar las columnas que no se utilizan
data = data.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
data


```

## Se crea el modelo de entrenamiento con los datos anteriores

```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report



# renombrar las columnas
data.columns = ['label', 'message']

# Convertir las etiquetas (ham -> 0, spam -> 1)
data['label'] = data['label'].map({'ham': 0, 'spam': 1})

# Eliminar filas con valores NaN en la columna 'label'
data = data.dropna(subset=['label'])

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(data['message'], data['label'], test_size=0.4, random_state=42) #se almacena el 60% de los datos para entrenamiento y el 40% para prueba


####### Aqui se entrenara el modelo
#Converir el texto en una representación numérica
vectorizer = CountVectorizer(stop_words='english')
X_train_vectorized = vectorizer.fit_transform(X_train) # se ajusta el vectorizador a los datos de entrenamiento y transforma los datos de entrenamiento en una representación numérica
X_test_vectorized = vectorizer.transform(X_test) # se transforma los datos de prueba en una representación numérica utilizando el mismo vectorizador ajustado a los datos de entrenamiento

# se utiliza un modelo de Naive Bayes multinomial para clasificar los mensajes
# este funciona bien con datos de texto y es eficiente para conjuntos de datos grandes
# utiliza el recuento de palabras como características de entrada, necesita un array de recuentos de palabras como entrada
# y devuelve una probabilidad de que el mensaje sea spam o ham
model = MultinomialNB()
model.fit(X_train_vectorized, y_train)
y_pred = model.predict(X_test_vectorized)

accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
classification = classification_report(y_test, y_pred)

print("Accuracy: {:.2f}%".format(accuracy * 100))
print("Confusion matrix:\n", confusion)
print("Classification report:\n", classification)


# mensajes para probar el modelo entrenado
messages = [
    "Congratulations! You've been selected as a winner. Call 123456 to claim your prize.",
    "Hey, what are you up to? Do you want to hang out later?",
    "Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question (std txt rate) T&C's apply 08452810075over18's",
    "URGENT! Your mobile No 1234 was awarded a Prize",
    "Hi, how are you? I hope you're doing well.",
    "Hi, i want ur help with something can u plz call me asap",
]



# Convertir los mensajes en una representación numérica
messages_vectorized = vectorizer.transform(messages)

predictions = model.predict(messages_vectorized)
for message, prediction in zip(messages, predictions):
    print('Message:', message)
    print('Prediction:', 'Spam' if prediction == 1 else 'Ham')


    
```

## Utilizar el teorema de la probabilidad junto con el teorema de Bayes para calcular la probabildad de que un correo sea spam o no

```python
# Usaremos el teorema de Bayes
# P(Spam|mensaje) = P(mensaje|Spam) * P(Spam) / P(mensaje)
# P(Spam) = es la proporcion de corres spam en el dataset
# P(mensaje) = es la probabilidad de que un mensaje ocurra
# P(mensaje|Spam) = es la probabilidad de que el mensaje sea spam
import numpy as np
import pandas as pd
from collections import Counter

data1 = pd.read_csv('./Dataset/spam.csv', encoding='ISO-8859-1') 
data1 = data1.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)

# Rename columns
data1.columns = ['label', 'message']
data1['label'] = data1['label'].map({'ham': 0, 'spam': 1})

P_spam = np.mean(data1['label']) # Proporción de correos spam en el dataset es igual a cantidadSpam / cantidadTotal
P_ham = 1 - P_spam # Proporción de correos no spam en el dataset

# Contar la frecuencia de palabras en los correos spam y no spam
frecuencia_palabras_spam = Counter()
frecuencia_palabras_nospam = Counter()
print(f"Proporción de spam: {P_spam:.4f}", f"Proporción de nospam: {P_ham:.4f}")

# contar aquellas palabras que aparecen en spam y no spam
for _, row in data1.iterrows():
    words = row['message'].lower().split()
    if row['label'] == 1:  # spam
        frecuencia_palabras_spam.update(words)
    else:  # ham
        frecuencia_palabras_nospam.update(words)
        
print(f"Total Palabras spam: {len(frecuencia_palabras_spam)}")
print(f"Total Palabras no spam: {len(frecuencia_palabras_nospam)}")

#calculamos el total e palabras spam y no spam
total_palabras_spam = sum(frecuencia_palabras_spam.values())
total_palabras_nospam = sum(frecuencia_palabras_nospam.values())

# Funcion del teorema de Bayes
def Bayes (message):
    words = message.lower().split()
    P_mensaje_spam = 1
    P_mensaje_nospam = 1

    for word in words:
        # Probabilidad de que la palabra aparezca en spam
        P_palabra_ap_spam = (frecuencia_palabras_spam[word] + 1) / (total_palabras_spam + len(frecuencia_palabras_spam))
        P_mensaje_spam *= P_palabra_ap_spam

        # Probabilidad de que la palabra aparezca en ham
        P_palabrea_ap_nospam = (frecuencia_palabras_nospam[word] + 1) / (total_palabras_nospam + len(frecuencia_palabras_nospam))
        P_mensaje_nospam *= P_palabrea_ap_nospam

    # Aplicar el teorema de Bayes
    P_spam_dado_mensaje = (P_mensaje_spam * P_spam) / ((P_mensaje_spam * P_spam) + (P_mensaje_nospam * P_ham))

    return P_spam_dado_mensaje

mensaje_prueba = "Congratulations! You've been selected as a winner. how are you? Yo are the winner"
print(f"Probabilidad de spam: {Bayes(mensaje_prueba):.4f}")
print(f"Probabilidad de spam (%): {Bayes(mensaje_prueba) * 100:.2f}%")
```
