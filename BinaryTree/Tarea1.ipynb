```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.left = None
        self.right = None
        
class Tree:
    def __init__(self):
        self.raiz = None
        
    def insertarNodo(self,dato):
        if self.raiz == None:
            self.raiz = Nodo(dato)
        else:
            self.insertar(self.raiz, dato)
            
    def insertar(self, nodo, dato):
        if dato < nodo.valor:
            if nodo.left == None:
                nodo.left = Nodo(dato)
            else:
                self.insertar(nodo.left, dato)
        else:
            if nodo.right == None:
                nodo.right = Nodo(dato)
            else:
                self.insertar(nodo.right, dato)

    def preOrden(self, nodo):
        if nodo != None:
            print(nodo.valor)
            self.preOrden(nodo.left)
            self.preOrden(nodo.right)
            
    def inOrden(self, nodo):
        if nodo != None:
            self.inOrden(nodo.left)
            print(nodo.valor)
            self.inOrden(nodo.right)
            
    def postOrden(self, nodo):
        if nodo != None:
            self.postOrden(nodo.left)
            self.postOrden(nodo.right)
            print(nodo.valor)

def main():
    arbol = Tree()
    n = int(input("Ingrese la cantidad de nodos: "))
    for i in range(n):
        dato = int(input("Ingrese el valor del nodo: "))
        arbol.insertarNodo(dato)
        
    print("PreOrden")
    arbol.preOrden(arbol.raiz)
    print("InOrden")
    arbol.inOrden(arbol.raiz)
    print("PostOrden")
    arbol.postOrden(arbol.raiz)
    dato = int(input("Ingrese el valor a buscar: "))
    print(arbol.buscar(arbol.raiz, dato))
    
main()
```
